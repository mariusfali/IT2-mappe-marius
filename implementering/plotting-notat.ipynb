```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 3*x**2
```

```python
x = []
y = []
for i in range(20):
    x.append(i)
    y.append(f(i))
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 200, 10000)
y = f(x)
```

```python
plt.style.use("bmh")
plt.plot(x, y)
plt.title("$3x^2$") # Dollar tegnet betyr "mattespråk", ^2 blir bedre vist
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```
